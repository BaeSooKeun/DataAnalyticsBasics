```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xlim(0,3)
plt.ylim(5,7)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xlim(left = 10) 
plt.ylim(top=10)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xticks([1,2,3,4],['A','B','C','D'])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.xticks([1,2,3,4],['A','B','C','D'])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.scatter([1,2,3,4],[8,7,6,5], label='x=3') #   .
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.scatter([1,2,3,4],[8,7,6,5], label='x=3') #   .
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc = 'center')
```

```python
plt.figure(figsize=(3,3))
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
```

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6], label= 'x1')
plt.plot([1,2,3],[6,5,4], label= 'x2')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],color='red' ,label= 'x1')
plt.plot([1,2,3],[6,5,4],color='blue',label= 'x2')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],color='red',linewidth=8,linestyle='dotted',label= 'x1')
plt.plot([1,2,3],[6,5,4],color='blue',linewidth=3,label= 'x2')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],marker='*',markersize=10,color='red' ,label= 'x1')
plt.plot([1,2,3],[6,5,4],color='blue',label= 'x2')
plt.legend()
```

```python
x1 = np.arange(1,10)
x2 = np.arange(1,30)
y1 = x1*2
y2 = x2*3
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,7))

#   !
ax[0].plot(x1,y1, color='red')
ax[1].bar(x2,y2, color='green')

ax[0].set_xlabel('ax[0] axis')
ax[1].set_xlabel('ax[1] axis')

ax[0].legend()
ax[1].legend()
```
